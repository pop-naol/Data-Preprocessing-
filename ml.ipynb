```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets    
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
file_path= r'D:\Downloads\breast_cancer\Breast_Cancer.csv'
ds= pd.read_csv(file_path)
ds.columns = ds.columns.str.strip()
ds.head()
```

```python
race = ds["Race"]
marital = ds["Marital Status"]
t_stage = ds["T Stage"]
n_stage = ds["N Stage"]
s_stage = ds["6th Stage"]
diff = ds["differentiate"]
a_stage = ds["A Stage"]
E_status = ds["Estrogen Status"]
E_status = E_status.replace({'Positive': 1, 'Negative': 0})
p_status = ds["Progesterone Status"]
F_status = ds["Status"]
F_status = F_status.replace({'Alive': 1, 'Dead': 0})

ds.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
xx = LabelEncoder()
race_t = xx.fit_transform(race)
marital_t = xx.fit_transform(marital)
t_stage_t = xx.fit_transform(t_stage)
n_stage_t = xx.fit_transform(n_stage)
s_stage_t = xx.fit_transform(s_stage)
diff_t = xx.fit_transform(diff)
a_stage_t = xx.fit_transform(a_stage)
E_status_t = xx.fit_transform(E_status)
p_status_t = xx.fit_transform(p_status)
F_status_t = xx.fit_transform(F_status)
```

```python
processed_df = pd.DataFrame({
    "Race": race_t,
    "Marital": marital_t,
    "T_Stage": t_stage_t,
    "N_Stage": n_stage_t,
    "Stage6": s_stage_t,
    "Differentiation": diff_t,
    "A_Stage": a_stage_t,
    "Estrogen_Status": E_status,
    "Progesterone_Status": p_status_t,
    "Final_Status": F_status
})
```

```python
le = LabelEncoder()

ds['Race'] = le.fit_transform(ds['Race'])
ds['Marital Status'] = le.fit_transform(ds['Marital Status'])
ds['T Stage'] = le.fit_transform(ds['T Stage'])
ds['N Stage'] = le.fit_transform(ds['N Stage'])
ds['6th Stage'] = le.fit_transform(ds['6th Stage'])
ds['differentiate'] = le.fit_transform(ds['differentiate'])
ds['A Stage'] = le.fit_transform(ds['A Stage'])
ds['Estrogen Status'] = ds['Estrogen Status'].replace({'Positive': 1, 'Negative': 0})
ds['Progesterone Status'] = le.fit_transform(ds['Progesterone Status'])
ds['Status'] = ds['Status'].replace({'Alive': 1, 'Dead': 0})
ds.head()
```

```python
y = ds['Status']
X = ds[['Age']]
plt.scatter(X, y, color='blue', )

```

```python
mode_2=KNeighborsClassifier(n_neighbors=3, metric='euclidean')
mode_2.fit(X,y)
y_pred = mode_2.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
mode_2=LogisticRegression()
mode_2.fit(X,y)
y_pred = mode_2.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_pred)
accuracy
```

```python
mode_2 = DecisionTreeClassifier()
mode_2.fit(X, y.ravel())  # flatten y for sklearn

y_pred = mode_2.predict(X)

accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)
```
